```python
def fitnessScoreForGivensubPaths(pathCost):
    pathCostwithSpaceFitness={}
    for key,value in pathCost.items():
        pathCostwithSpaceFitness[key]=value[0]*value[1]
    return pathCostwithSpaceFitness
```

```python
def deriveFitnessScoreForAllPath(possiblePath,pathCostwithSpaceFitness):
    for key,value in possiblePath.items():
        # print(key)
        key1=key.replace('N','')
        # print(key1)
        len_key = len(key1)
        pathCost1 = 0
        for i in range(len_key-1):
            path=key1[i:i+2]
            # print(path)
            # print("End")
            if path not in pathCostwithSpaceFitness.keys():
                pathfitness= 10000
            else:
                pathfitness=pathCostwithSpaceFitness[path]

            pathCost1 = pathCost1 + pathfitness
            #print(f'pathcost for {path} is {pathfitness}')
        #print(f'pathcost for {key} is {pathCost1}')
        possiblePath[key]=pathCost1
    return possiblePath
```

```python
import random
def sampleSelection(possiblePath,noOfItem):
    InitialPath=dict(random.sample(list(possiblePath.items()),noOfItem))
    return InitialPath
```

```python
def crossOverAfternthPoint(SelectedPath,n,n1):
    crossOverPath={}
    for i in range(0,len(SelectedPath),2):
        element1 = list(SelectedPath.keys())[i]
        #if after crossover in previous step same path came multiple time hence total no of paths got reduced from 6(even) to 5(odd)
        #in those cases selecting last element and first element for next crossover.
        if (i+1 == len(SelectedPath)):
            element2 = list(SelectedPath.keys())[0]
        else:
            element2 = list(SelectedPath.keys())[i+1]
        #print(element1)
        #print(element2)
        #print(element2[2:4])
        element1AfterCrossOver = element1[0:n]+element2[n:len(element2)]
        element2AfterCrossOver = element2[0:n]+element1[n:len(element1)]
        #print(element1AfterCrossOver)
        #print(element2AfterCrossOver)
        if (len(element1AfterCrossOver) != len(set(element1AfterCrossOver))) or (len(element2AfterCrossOver) != len(set(element2AfterCrossOver))):
            element1AfterCrossOver = element1[0:n1]+element2[n1:len(element2)]
            element2AfterCrossOver = element2[0:n1]+element1[n1:len(element1)]

        crossOverPath[element1AfterCrossOver]=0
        crossOverPath[element2AfterCrossOver]=0
    #print(SelectedPath)
    return crossOverPath
```
